### Question Two


```python
#setting constants
R = 0.083144 # Lbar/(molK), the ideal gas constant
a = 16.02 # barL^2/mol^2, parameter for acetone
b = 0.1124 # L/mol, parameter for acetone
n = 1 # mol, the amount of substance

#P and V Ranges
P = np.linspace(1,10,100)
V = np.linspace(10,30,100)

V_mesh, P_mesh = np.meshgrid(V, P) # taking the 2 1D arrays and creating a 2 2D martices
#Calculate Temperature
T = ((P_mesh + a * n**2 / V_mesh**2) * (V_mesh - n*b)) / (n*R)

#plot

plt.figure(figsize=(8,6))
temperature_plot = plt.contourf(V_mesh, P_mesh, T, 100, cmap='viridis', vmin=np.min(T), vmax=np.max(T)) # creates filled contour plots with 100 level, cmap defines color map, vmin and vmanx are the bounds for scaling

#Labels
plt.title('Temperature of One Mole Acetone')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.colorbar(temperature_plot)

#show
plt.show()

```

### Question 3

```python
#Generate random points
N = 10000 # number of random points to generate
points = np.random.rand(N, 2) # Generate points in 2D space
```

```python
#Distances

distances = np.sqrt(points[:, 0]**2 + points[:, 1]**2)
inside_circle = distances <= 1
```

```python
estimate_pi = 4 * np.sum(inside_circle) / N
```

```python
#Plot

plt.figure(figsize=(6, 6)) #1x1 square plot
plt.scatter(points[inside_circle, 0], points[inside_circle, 1], color='blue', s=1) # Points inside 
plt.scatter(points[~inside_circle, 0], points[~inside_circle, 1], color='red', s=1) # Points outside

plt.text(0.5, 0.5, f"  {estimate_pi:.4f}", fontsize=20, ha='center') # display pi estimate

plt.title('Estimate Pi')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.axis('equal') # Ensure the x and y axes have equal scales
plt.show() # Show the plot
```
