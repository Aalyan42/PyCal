```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#1a

#Load Data
df = pd.read_csv('GlobalLandTemperaturesbyState.csv')

#Filter Columns 
df = df[['dt', 'AverageTemperature', 'State']]

#Convert to datetime type
df['dt'] = pd.to_datetime(df['dt'])
#Filter years
df = df[df['dt'].dt.year > 2000]

#Filter states
df = df[df['State'].isin(['South Dakota', 'Wyoming', 'Nebraska'])]
```

```python
#1b

#Change table for average temperature
average_temp_df = df.groupby('dt')['AverageTemperature'].mean().reset_index()
average_temp_df.columns = ['Date', 'AverageTemperature']
```

```python
#1c

#Plot the data

plt.figure(figsize=(10,5))
temp_plot = plt.plot(average_temp_df['Date'], average_temp_df['AverageTemperature'], marker='o', linestyle='-')

plt.title('Average Temperature of Nebraska, South Dakota, and Wyoming after the year 2000')
plt.xlabel('Date')
plt.ylabel('Average Temperature')

plt.tight_layout()
plt.show()
```

```python
#1d

#Convert strings to numerical data
#already converted Date column to datetime

reference_date = average_temp_df['Date'].min()

average_temp_df['Days'] = (average_temp_df['Date'] - reference_date).dt.days
```

```python
#1e

#Model Equation
def temp_model(t, A, B, C, D):
    return A * np.cos(B * t + C) + D

guess = [10, 2 * np.pi / 365, 0, 20]
```

```python
#1f
x_vals = average_temp_df['Days']
y_vals = average_temp_df['AverageTemperature']

parameters, covariance = curve_fit(temp_model, x_vals, y_vals, p0=guess)
```

```python
#1g

#Replot the data

plt.figure(figsize=(10,5))
plt.scatter(average_temp_df['Days'], average_temp_df['AverageTemperature'], label='Data', color='blue')

fitted_temps = temp_model(average_temp_df['Days'], *parameters)

temp_plot2 = plt.plot(average_temp_df['Days'], fitted_temps, label='Fitted Curve', color='red')

plt.title('Fit of Model to Temperature Data')
plt.xlabel('Days since ' + str(reference_date.date()))
plt.ylabel('Average Temperature')
plt.legend
           
plt.tight_layout()
plt.show()
```

```python
#1h

#Errors

errors = np.sqrt(np.diag(covariance))

for i, (parameter, error) in enumerate(zip(parameters, errors)):
    print(f'Paramter {i+1}: {parameter:.4f} +/- {error:.4f}')
```

```python
#1i

#Print

equation = f"(T(t) = {parameters[0]:.4f} * cos{parameters[1]:.4f} * t + {parameters[2]:.4f} + {parameters[3]:.4f})"
print(f"The fitted equation is:\n{equation}")
```
